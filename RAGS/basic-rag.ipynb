```python
from langchain.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser

from langchain_community.document_loaders import JSONLoader
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS

# load document
loader = JSONLoader(file_path="knowledge_base.json", jq_schema=".[].content", text_content=True)
documents = loader.load()

# create embeddings
embedding_model = OpenAIEmbeddings()
embeddings = embedding_model.embed_documents([doc.page_content for doc in documents])
```

```python
embeddings
```

```python
# store in vector db
vector_db = FAISS.from_documents(documents, embedding=embedding_model)
# create retriever
retriever = vector_db.as_retriever(search_kwargs={"k": 5})

query = "What are the effects of climate change?"
retriever_results = retriever.get_relevant_documents(query)
retriever_results
```
